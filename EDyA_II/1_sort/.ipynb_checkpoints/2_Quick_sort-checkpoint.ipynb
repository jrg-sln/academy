## Quick sort

El método quick sort es uno de los más eficientes y veloces de los métodos de ordenación interna. También es conocido como método rápido o de ordenación por partición.

El algoritmo sigue los siguientes pasos:

- Se toma un elemento X de una posición cualquiera del conjunto.
- Se ubica al elemento X en la posición correcta del conjunto, de tal manera que todos los elementos que se encuentren a su izquierda sean menores o iguales a X y todos los elementos que se encuentren a su derecha sean mayores o iguales a X.
- Se repiten los pasos anteriores para los conjuntos de datos que se encuentran a la izquierda y a la derecha de la posición de X en el conjunto.

El proceso termina cuando todos los elementos se encuentran ordenados.


```python
# https://www.geeksforgeeks.org/sorting-algorithms/

%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
TAM = 200

def partition(intList, low, high):
    i = low-1
    pivot = intList[high]
    for j in range(low , high):
        if   intList[j] <= pivot:
            i = i+1
            intList[i],intList[j] = intList[j],intList[i]
 
    intList[i+1],intList[high] = intList[high],intList[i+1]
    return (i+1)
```

```python
def quick_sort(intList, low, high):
    if low < high:
        pi = partition(intList, low, high)
        quick_sort(intList, low, pi-1)
        quick_sort(intList, pi+1, high)
```

```python
l = []
for i in range(1, 21, 1):
    l.append(random.randint(1, 100))

print(l)
quick_sort(l, 0, len(l)-1)
print (l)
```

```python
time = 0

def partition_graph(arr, low, high):
    global time
    i = low-1
    pivot = arr[high]
    for j in range(low , high):
        time += 1
        if   arr[j] <= pivot:
            i = i+1
            arr[i],arr[j] = arr[j],arr[i]
 
    arr[i+1],arr[high] = arr[high],arr[i+1]
    return i+1


def quick_sort_graph(arr, low, high):
    global time
    if low < high:
        time += 1
        pi = partition_graph(arr, low, high)
        quick_sort_graph(arr, low, pi-1)
        quick_sort_graph(arr, pi+1, high)
```

```python
x = list(range(1,TAM + 1,1))
y_omega = []
y_efedeene = []
y_omicron = []
L = []

for num in x:
    # Random list
    L = []
    for i in range(1, num+1, 1):
        L.append(random.randint(1, TAM*10))
    
    # average case
    time = 0
    quick_sort_graph(L, 0, len(L)-1)
    y_efedeene.append(time)
    
    # best case
    L.sort()
    time = 0
    quick_sort_graph(L, 0, len(L)-1)
    y_omega.append(time)
    
    # worst case
    L.sort()
    L.reverse()
    time = 0
    quick_sort_graph(L, 0, len(L)-1)
    y_omicron.append(time)


#print(y_omega)
#print(y_efedeene)
#print(y_omicron)
```

```python
fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_omega, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["quick sort"])
plt.title('quick sort (best case)')
plt.show()

fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_efedeene, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["quick sort"])
plt.title('quick sort (average case)')
plt.show()

fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_omicron, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["quick sort"])
plt.title('quick sort (worst case)')
plt.show()

#fig.savefig("buble_sort.png")
```
