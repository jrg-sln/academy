```python
import random

#Funcion generadora de listas aleatorias
def randomList(n):
    lista = [0]  * n
    for i in range(n):
        lista[i] = int(random.random()*1000)
    return lista
```

```python
#Funcion generadora de listas aleatorias
def randomList2(n):
    lista = [0]  * n
    lista[0] = 0
    i = 1
    while i<n:
        lista[i] = int(random.random()*1000)
        i=i+1
    return lista
```

```python
#Esta línea se ocupa para que las gráficas que se generen queden embebidas dentro de la página
%pylab inline  

#Importando las bibliotecas
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Función que realiza los intercambios.
def swap(A, x, y):
    tmp = A[x]
    A[x] = A[y]
    A[y] = tmp

#Función que divide al conjunto usando un pivote.
def split(A, p, r, time):
    x = A[(p+r)//2]     #Pivote 
    #x = A[r]
    i = p-1
    for j in range(p, r):
        time += 1
        if(A[j]<=x):
            i = i+1
            swap(A, i, j)
    swap(A, i+1, r)
    return i+1, time

#Función de ordenamiento QuickSort
def QuickSort(A, p, r, time):
    time += 1
    if(p < r):
        Q = split(A, p, r, time)
        q = Q[0]
        time = Q[1]
        time = QuickSort(A, p, q-1, time)
        time = QuickSort(A, q+1, r, time)
    return time
```

```python
#Prueba de QuickSort
Pr1 = randomList(20)
print('Lista original: ', Pr1)
print('Tiempo de ejecución: ', QuickSort(Pr1, 0, len(Pr1)-1, 0))
print('Lista ordenada: ', Pr1)
```

```python
#Datos de entrada
x = list(range(1, 500, 1))
y = []
for num in x:
    L = randomList(num)
    L.sort()
    #L = list(range(1, num+1 ,1))
    y.append(QuickSort(L, 0, len(L)-1, 0))
```

```python
fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y, marker="o",color="r", linestyle='None')

ax.grid(True)
ax.set_xlabel('X')  #Etiqueta del eje x
ax.set_ylabel('Y')  #Etiqueta del eje y
ax.grid(True)
ax.legend(["QuickSort Mejor Caso"])

plt.title('Puntos')
plt.show()

fig.savefig("P2gráfica1.png") #Guardando la gráfica
```

```python
import math

def left(i):
    return 2*i

def right(i):
    return 2*i+1

def swap(A, x, y):
    tmp = A[x]
    A[x] = A[y]
    A[y] = tmp

#Función que mantiene el heap máximo
def MaxHeapify(A, i, heapSize, time):
    time += 1
    L = left(i)
    R = right(i)
    if(L <= heapSize and A[L]>A[i]):
        posMax = L
    else:
        posMax = i
        
    if(R <= heapSize and A[R]>A[posMax]):
        posMax = R
    if(posMax != i):
        swap(A, i, posMax)
        time = MaxHeapify(A, posMax, heapSize, time+1)
    return time
        
#Función que crea al heap máximo
def BuildMaxHeapIni(A, heapSize, time):
    for i in range(math.ceil(heapSize/2)-1, 0, -1):
        time = MaxHeapify(A, i, heapSize, time)
    return time

#Función de ordenamiento HeapSort
def heapSort(A, heapSize, time):
    time += 1
    time = BuildMaxHeapIni(A, heapSize, time+1)
    for i in range(len(A[1:]), 1, -1):
        swap(A, 1, i)
        heapSize -= 1
        time = MaxHeapify(A, 1, heapSize, time+1)
    return time
```

```python
#Prueba de HeapSort
Pr2 = randomList(20)
print('Lista original: ', Pr2)
print('Tiempo de ejecución: ', heapSort(Pr2,len(Pr2)-1, 0))
print('Lista ordenada: ', Pr2)
```
