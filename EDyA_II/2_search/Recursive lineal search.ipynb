## Recursive lineal search

La búsqueda lineal, también llamada secuencial, es el método de búsqueda más sencillo, consiste en revisar elemento tras elemento hasta encontrar el dato buscado o llegar al final del conjunto de datos.
Este método se puede aplicar a estructuras de datos tanto ordenadas como desordenadas. Se suele utilizar cuando el conjunto de datos es pequeño.

```python
%pylab inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import random
```

```python
def lineal_search_recursive(entList, entValue, entCont):
    if entCont == len(entList):
        return -1
    
    if entList[entCont] == entValue:
        return entCont
    
    entCont += 1
    return lineal_search_recursive(entList, entValue, entCont)
```

```python
TAM = 21
val = 50

l = []
for i in range(1, TAM, 1):
    l.append(random.randint(1, 100))

print(l)
res = lineal_search_recursive(l, val, 0)

if res == -1:
    print('La llave', val,'no se encuentra en el conjunto')
else:
    print('El valor', val,'se encuentra en la posición', res)

l.sort()
res = lineal_search_recursive(l, val, 0)
print(l)
if res == -1:
    print('La llave', val,'no se encuentra en el conjunto')
else:
    print('El valor', val,'se encuentra en la posición', res)
```

```python
class Node:
    def __init__(self, number, name, lastname, email, sex):
        self.identifier = number
        self.name = name
        self.last_name = lastname
        self.email = email
        self.sex = sex[0]
        
    def __str__(self):
        return self.name + ":" + self.email
```

```python
def lineal_node_search_recursive(nodeList, nodeValue, entCont):
    if entCont == len(nodeList):
        return -1
    
    if nodeList[entCont].name == nodeValue.name:
        return entCont
    
    entCont += 1
    return lineal_node_search_recursive(nodeList, nodeValue, entCont)
```

```python
nodeList = []
file = open("DATA_10.csv", "r")
for line in file:
    fields = line.split(",")
    nodeList.append(Node(fields[0], fields[1], fields[2], fields[3], fields[4]))
        
print("List")
for node in nodeList:
    print(node)
    
value = nodeList[random.randint(0, len(nodeList))]
print("\nSearch ", value.name)
print(lineal_node_search_recursive(nodeList, value, 0))
```

```python
def lineal_node_search_recursive_graph(nodeList, nodeValue, entCont):
    global cont
    cont += 1
    if entCont == len(nodeList):
        return -1
    
    if nodeList[entCont].name == nodeValue.name:
        return entCont
    
    entCont += 1
    return lineal_node_search_recursive_graph(nodeList, nodeValue, entCont)
```

```python
TAM = 1001
x = list(range(1,TAM,1))
y_omega = []
y_efedeene = []
y_omicron = []
L = []
cont = 0
for num in x:
    iter = 1
    file = open("DATA_10000.csv", "r")
    L = []
    for line in file:
        fields = line.split(",")
        L.append(Node(fields[0], fields[1], fields[2], fields[3], fields[4]))
        
        if iter == num:
            break
        iter += 1
    
    # average case
    value = L[random.randint(0, len(L)-1)]
    cont = 0
    lineal_node_search_recursive_graph(L, value, 0)
    y_efedeene.append(cont)

    # best case
    value = L[0]
    cont = 0
    lineal_node_search_recursive_graph(L, value, 0)
    y_omega.append(cont)

    # worst case
    value = Node(0, "zzzzzzzzzzzz", "zzzzzzzzzzzz", "zzzzzzzzzzzz", "zzzzzzzzzzzz")
    cont = 0
    lineal_node_search_recursive_graph(L, value, 0)
    y_omicron.append(cont)
```

```python
fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_omega, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["Lineal iterative search"])
plt.title('Lineal iterative search (best case)')
plt.show()

fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_efedeene, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["Lineal iterative search"])
plt.title('Lineal iterative search (average case)')
plt.show()

fig, ax = plt.subplots(facecolor='w', edgecolor='k')
ax.plot(x, y_omicron, marker="o",color="b", linestyle='None')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.grid(True)
ax.legend(["Lineal iterative search"])
plt.title('Lineal iterative search (worst case)')
plt.show()
```
